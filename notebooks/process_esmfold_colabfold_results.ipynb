```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
esm_file = "../esmfold/esmfold_adeno_cr1_cmv_rl11_elephantidbeta_ee50_results_sorted_by_plddt.csv"
esm_df = pd.read_csv(esm_file)
colab_file = "../colabfold/colabfold_adeno_cr1_cmv_rl11_elephantidbeta_ee50_results_sorted_by_plddt.csv"
colab_df = pd.read_csv(colab_file)
```

```python
# Merge the two dataframes based on protein id
merged_df = pd.merge(esm_df, colab_df, on='protein_id', suffixes=('_esm', '_colab'))
```

```python
merged_df
```

```python
# Plot the correlation of pLDDT scores with seaborn

# plot the scatterplot
plt.figure(figsize=(6,6))
sns.scatterplot(data=merged_df, x="pLDDT_esm", y="pLDDT_colab", color='black', alpha=0.5, s=35)
plt.xlabel("ESMFold pLDDT")
plt.ylabel("ColabFold pLDDT")

# add a diagonal line
#plt.savefig("plddt_correlation.png")
x = np.linspace(0, 100, 100)
plt.plot(x, x, color='black', linestyle='-', linewidth=0.75)

plt.show()
```
